```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data for sentiment distribution
sentiments = ["Positive", "Negative", "Neutral", "Positive", "Neutral", "Negative", "Positive", "Neutral"]
# Create a DataFrame to count occurrences
data = pd.DataFrame({"Sentiment": sentiments})

# Count occurrences of each sentiment
sentiment_counts = data["Sentiment"].value_counts()

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.2f%%')

# Add a title
plt.title('Distribution of Sentiment')

# Show the plot
plt.show()
```

 here, the confusion matrix of sentiment analysis. In a sentiment analysis problem, the confusion matrix shows how many sentiments were correctly predicted versus how many were misclassified.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report

# Example: Actual and predicted labels (replace with your actual data)
y_test = [1, 0, 2, 1, 0, 1, 2, 0, 1, 2]  # Actual sentiments (e.g., 0: Negative, 1: Positive, 2: Neutral)
y_pred = [1, 0, 2, 0, 0, 1, 2, 1, 1, 2]   # Predicted sentiments

# Step 1: Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Step 2: Precision (for each class: Positive, Negative, Neutral)
precision = precision_score(y_test, y_pred, average=None)  # average=None returns per class precision
print("Precision (per class):", precision)

# Step 3: Recall (for each class: Positive, Negative, Neutral)
recall = recall_score(y_test, y_pred, average=None)  # average=None returns per class recall
print("Recall (per class):", recall)

# Step 4: F1-Score (for each class: Positive, Negative, Neutral)
f1 = f1_score(y_test, y_pred, average=None)  # average=None returns per class F1-Score
print("F1-Score (per class):", f1)

# Step 5: Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# Display Confusion Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Negative", "Positive", "Neutral"], yticklabels=["Negative", "Positive", "Neutral"])
plt.title('Confusion Matrix for Sentiment Analysis')
plt.xlabel('Predicted Sentiment')
plt.ylabel('True Sentiment')
plt.show()

# Step 6: Classification Report
# Provides precision, recall, f1-score, and support for each class
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=["Negative", "Positive", "Neutral"]))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data for sentiments per category
categories = ['Category 1', 'Category 2', 'Category 3']
positive = [10, 15, 5]
negative = [3, 2, 5]
neutral = [7, 5, 12]

# Create stacked bar chart
barWidth = 0.5
bars = [positive, negative, neutral]

# Set position of bars on X axis
r = np.arange(len(categories))

plt.bar(r, positive, color='green', width=barWidth, label='Positive')
plt.bar(r, negative, color='red', width=barWidth, bottom=positive, label='Negative')
plt.bar(r, neutral, color='blue', width=barWidth, bottom=np.array(positive) + np.array(negative), label='Neutral')

plt.title('Sentiment Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(r, categories)
plt.legend()
plt.show()
```
